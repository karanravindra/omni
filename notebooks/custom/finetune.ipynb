```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
tokenizer = AutoTokenizer.from_pretrained("unsloth/gemma-3-270m-it-GGUF")
model = AutoModelForCausalLM.from_pretrained("unsloth/gemma-3-270m-it-GGUF", device_map=device, torch_dtype=torch.bfloat16)
```

```python
messages = [
    {"role": "user", "content": "Who are you?"},
]
inputs = tokenizer.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)
```

```python
outputs = model.generate(**inputs, max_new_tokens=40)
print(tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```
