```python
from datasets import load_dataset
from ollama import Client
from tqdm import tqdm
```

```python
SPLIT = "train"
DATASET = "HuggingFaceH4/ultrachat_200k"

ds = load_dataset(DATASET, split="test_sft")

print(f"Dataset: {DATASET}, Split: {SPLIT}, Size: {len(ds):,}")

num_exact_duplicates = len(ds) - len(set(ds["prompt"]))
print(f"Number of exact duplicate prompts: {num_exact_duplicates:,}")
```

```python
ollama = Client("https://ollama.karanravindra.com")
```

```python
def encode_batch(prompts: list[str]):
    print(prompts)
    return ollama.embed(model="embeddinggemma", input=prompts).embeddings

ds = ds.map(
    lambda x: {"embedding": encode_batch(x["prompt"])},
    batched=True,
    batch_size=32
)
```

```python
def print_conversation(messages: list[dict[str, str]]):
    for message in messages:
        role = message['role']
        content = message['content']
        print("=" * 80)
        print(f"{role.upper()}:\n{content}\n")
```
